판다스에서 데이터 프레임을 특정 행 또는 열 기준으로 나누고 싶을 때 주로 loc와 iloc을 주로 이용한다.

loc: '변수명'을 기준으로 분리한다.

iloc: '인덱스 번호'로 분류한다.

예시)
- 기본 형태: iloc[ : , : ]
    - 행을 불러오기:
        - iloc[0]: 첫 번째 행만
        - iloc[1]: 두 번째 행만
    - 열을 불러오기:
        - iloc[ : , a:b]  
        - iloc[ : , a]  
        - iloc[ : , :b]

```python
# 데이터 정렬
# sort_values() 함수 사용
import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table', header = 0, index_col = 0) 
summer = df[1].iloc[:,:5]
summer.columns = ['경기수', '금', '은', '동', '계']
summer.sort_values('금', ascending = False) # 금을 기준으로 내림차순
```

### 데이터 프레임 기초
2차원 데이터 형태의 데이터 프레임은 행과 열이 있고, 행을 구분해주는 인덱스와 열을 구분해주는 칼럼이 있다.  
그래서 별도로 지정해주지 않으면 인덱스는 리스트처럼 정수로 설정이 되고 한 번 설정된 인덱스는 변경되지 않는다.

```python
# 날짜 형태로 된 8개의 인덱스 생성
import pandas as pd
index = pd.date_range('1/1/2000', periods=8)
print(index)
```

```python
# np.random.rand(a, b): a행 b열의 2차원 배열 생성(데이터는 0~1사이의 실수)
np.random.rand(8, 3)
```

```python
# 인덱스 출력
list('ABC')
```

```python
import numpy as np
df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC')) # 데이터, 행 이름, 열 이름 지정
df
```

```python
# 데이터 프레임에서 특정 행이나 열을 선택
# 1차원 배열 형태의 데이터로 이루어진 시리즈(Series)라는 데이터 구조 형태로 표현된다.

import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8, 3), index = index, columns = list('ABC'))
print(df['B'])
```

```python
# 데이터 프레임에 마스크 기능 사용
# 마스크: 특정한 조건을 만족하는지에 따라 참, 거시을 반환하여 원하는 데이터를 골라내는 데 유용함

import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8, 3), index = index, columns = list('ABC'))
print(df['B'] > 0.4)
```

```python
# B열의 값이 0.4보다 큰 값이라는 조건이 True인 데이터들만 이루어진 데이터 프레임 생성
df2 = df[df['B']>0.4]
df2
```

```python
# 행과 열의 데이터를 뒤집기(T = Transpose)
df2.T
```

2차원 배열 형태의 데이터 프레임 연산

```python
# 행 방향 축으로 새로운 칼럼 생성
import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8, 3), index = index, columns = list('ABC'))
df['D'] = df['A'] / df['B']    # A열의 값을 B열의 값으로 나눈 값을 D열에 저장
df
```

```python
# 열 방향 축으로 새로운 칼럼 생성
import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8, 3), index = index, columns = list('ABC'))
df['D'] = df['A'] / df['B']    
df['E'] = np.sum(df, axis=1)   # 행 우선 계산 값을 E에 저장
df.head()                      # 맨 위 5개 행만 추출
```

axis가 0이면 행 방향, axis가 1이면 열 방향을 축으로 삼아 계산한다.

```python
# 전체 데이터에 대해서 열 우선 계산
## 전체 데이터를 A열의 값으로 뺀다.
import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8, 3), index = index, columns = list('ABC'))
df['D'] = df['A'] / df['B']    
df['E'] = np.sum(df, axis=1)   
df = df.sub(df['A'], axis=0)    # A열 데이터를 기준으로 열 우선 계산
df.head()           

# A열 값을 기준으로 전체 데이터에 대한 뺄셈이 이루어졌기 때문에 A열 값은 모두 0이된다.
# B열 값도 기존의 값에서 A열 값만큼 뺀 값이 저장된 것을 확인할 수 있다.
```

```python
import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8, 3), index = index, columns = list('ABC'))
df['D'] = df['A'] / df['B']    
df['E'] = np.sum(df, axis=1)   
df = df.sub(df['A'], axis=0)
df = df.div(df['C'], axis=0)    # C열 데이터를 기준으로 열 우선 계산
df.to_csv('test.csv')
df.head()           
```

### pandas로 인구 구조 분석하기

```python
# 1. 데이터 읽어오기
import pandas as pd
df = pd.read_csv('age14.csv', encoding = 'cp949', index_col=0)
df.head()
```

```python
import pandas as pd
df = pd.read_csv('age14.csv', encoding = 'cp949', index_col=0)
df = df.div(df['총인구수'], axis=0) # 전체 데이터를 총인구수로 나눠서 비율로 변환
del df['총인구수'], df['연령구간인구수'] # 총인구수, 연령구간인구수 열 삭제
```

```python
# 2, 3. 궁금한 지역 이름 입력받고 해당지역의 인구 구조 저장하기
# df.index.str.contains() 함수: 데이터 프레임의 인덱스 문자열에 원하는 문자열이 포함된 행을 찾는다.
name = input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name) # 2. 지역 이름 입력
df2 = df[a] # 3. 해당 행을 찾아서 해당 지역의 인구 구조를 저장
df2
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
df2.T.plot()
plt.show()
```

```python
# 4. 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 시각화하기
import numpy as np

# (1) 전체 df에서 df2의 첫 번째 행을 각각 뺀다.
x = df.sub(df2.iloc[0], axis = 1)
# (2) 뺀 값의 제곱 값을 모두 더한다.
y = np.power(x, 2)
z = y.sum(axis=1)
print(z)
```

```python
i = z.sort_values().index[:5] # (3) 그 차이가 가장 작은 지역 5곳을 찾는다.
print(i)
```

```python
df.loc[i].T.plot() # (4) 결과를 꺾은선 그래프로 보여준다.
plt.show()

# df에서 i(변수명)에 해당하는 데이터를 행, 열 전치해서 꺾은선 그래프로 작성한다.
```
