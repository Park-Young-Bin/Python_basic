```python
import csv
f =open('seoul.csv')        # CSV 모듈 불러오기
data = csv.reader(f)        # seoul.csv 파일 읽기 모드로 불러오기
header =next(data)
max_temp =-999     # 최고 기온 값을 저장할 변수 초기화
max_date =''       # 최고 기온이 가장 높았던 날짜를 저장할 변수

for row in data :
    if row[4] =='' :     # 만약 최고 기온 값이 비어있다면...
        row[4] =-999     # -999를 넣어 빈 문자열이 있던 자리라고 표시
    row[4] = float(row[4])
    if max_temp < row[4] :     # 만약 지금까지 최고 기온보다 높아면 업데이터
        max_date = row[2]
        max_temp = row[4]
f.close()
print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은',max_date+'로, ', max_temp, '도 였습니다.')
```
