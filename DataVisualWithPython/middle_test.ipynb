```python
# (2) 서울시 외국인인구수를 꺽은선 그래프로 나타내시오.

import csv
f=open('seoul.csv')
data=csv.reader(f)
header = next(data)

name = input('알고 싶은 지역의 이름을 입력해주세요 : ')
result = []

for row in data:
    if name in row[2]:
        row[3:] = map(int, row[3:])
        result.append(row[5])
        
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+ ' 지역의 외국인인구수')
plt.plot(result)
plt.show()
```

```python
# (3) 종로구의 18년 5월 내국인생활 인구수와 19년 5월 내국인생활 인구수를 파이차트로 보이시오.

import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
eight = 0
nine = 0
size = []
name = input('알고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
    if name in row[2]:
        for i in range(3,6):
            row[i] = int(row[i])
            date = row[0]
            if date[:6] == '201805':
                eight = row[4]
            if date[:6] == '201905':
                nine = row[4]

size.append(eight)
size.append(nine)
print(size)

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 지역의 18년과 19년 5월 내국인생활인구수 비교')
plt.pie(size, colors = ['skyblue', 'hotpink'], labels = ['18년 5월', '19년 5월'], autopct = '%.1f%%', startangle = 90)
plt.axis('equal')
plt.show()
```

```python
# (2) 송파구 지역의 총생활인구수와 내국인생활인구수를 꺾은선 그래프로 나타내시오.

import csv
f=open('seoul.csv')
data=csv.reader(f)
name = input('찾고 싶은 지역의 이름을 알려주세요 :')
tot = []
kor = []

for row in data:
    if name in row[2]:
        tot.append(int(row[3]))
        kor.append(int(row[4]))
        
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 지역의 총생활인구수와 내국인생활인구수')
plt.plot(tot, color = 'r', label='총생활인구수')
plt.plot(kor, color = 'b' ,label='내국인생활인구수')
plt.legend()
plt.show()        
```

```python
# (3) 2018년의 송파구의 총생활인구수와 내국인생활인구수를 파이차트로 나타내시오.

import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
tot = 0
kor = 0
size = []

for row in data:
    if name in row[2] and row[0][:4] == '2018':
        tot = tot + int(row[3])
        kor = kor + int(row[4])
size.append(tot)
size.append(kor)
print(size)

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 지역의 총생활인구수와 내국인생활인구수')
label = ['총생활인구수', '내국인생활인구수']
color = ['skyblue', 'hotpink']
plt.pie(size, labels=label, colors=color, autopct = '%.1f%%', startangle=90)
plt.show()
```
